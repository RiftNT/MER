```python
import librosa
import numpy as np
import glob
import os
```

```python
def extract_mel_spectrogram(audio_path, output_path, n_mels=128):
    # load audio file
    y, sr = librosa.load(audio_path, sr=None)
    
    # extract melspectrogram of audio
    mel_spectrogram = librosa.feature.melspectrogram(y=y, sr=sr, n_mels=n_mels)
    
    # convert to db scale
    mel_spectrogram_db = librosa.power_to_db(mel_spectrogram, ref=np.max)
    
    # save spectrogram
    npy_filename = os.path.splitext(os.path.basename(audio_path))[0] + '.npy'
    npy_path = os.path.join(output_path, npy_filename)
    np.save(npy_path, mel_spectrogram_db)
    print(f"Mel spectrogram saved as: {npy_path}")
```

```python
def process_directory(input_dir, output_dir, n_mels=128):
    # get all audio files
    audio_files = glob.glob(os.path.join(input_dir, '*.mp3'))
    
    # process all audio files
    for audio_file in audio_files:
        extract_mel_spectrogram(audio_file, output_dir, n_mels)
```

```python
input_directory = 'data/audio/'
output_directory = 'data/spectrogram/'
process_directory(input_directory, output_directory)
```
